```python
# dataset-Algerian Forest Fire dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv("forestfire_cleaned_dataset",header=1)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
df
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
df.head()
```

```python
df.isnull().sum()
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.rename(columns={
    '1':"Day",
    '6':"Month",
    '2012':"Year",
    '29':"Temp",
    '57':"Rh",
    '18':"WS",
    '0.0':"Rain",
    '65.7':"FFMC",
    '3.4':"DMC",
    '7.6':"DC",
    '1.3':"ISI",
    '3.4.1':"BWI",
    '0.5':"FWI",
    'not fire':"classes",
    "Region":"Region"}, inplace=True
)
df.head()
```

```python
df.columns
```

```python
df[['Day', 'Month', 'Year', 'Temp', 'Rh', 'WS']]=df[['Day', 'Month', 'Year', 'Temp', 'Rh', 'WS']].astype(int)
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='classes':
        df[i]=df[i].astype(float)
```

```python
df.to_csv('forestfire_cleaned_dataset.csv',index=False)
```

```python
#EDA
dfcopy=df.drop(['Day','Month','Year'],axis=1)
dfcopy.head()
```

```python
dfcopy['classes'].value_counts
```

```python
dfcopy['classes']=np.where(dfcopy["classes"].str.contains("not fire"),0,1)
```

```python
# independent and dependent features
# FWI-fire weather index
x=dfcopy.drop('FWI',axis=1)
y=dfcopy['FWI']
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
# performing feature selection
x_train.corr()
# use of correlation- correlation tells us what features are how closley related to each other so that
# the features that are highly correlated we can drop one of them
```

```python
plt.figure(figsize=(12,10))
corr=x_train.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
def corre(dataset, threshold):
    colcorr=set()
    corrmat=dataset.corr()
    for i in range(len(corrmat.columns)):
        for j in range(i):
            if abs(corrmat.iloc[i,j])>threshold:
                colname=corrmat.columns[i]
                colcorr.add(colname)
    return colcorr
```

```python
corrfeatures=corre(x_train,0.85)
corrfeatures
```

```python
# any column that has highly correlation>threshold will get dropped
x_train.drop(corrfeatures,axis=1,inplace=True)
x_test.drop(corrfeatures,axis=1,inplace=True)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# model training
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
print(score)
plt.scatter(y_test,y_pred)
plt.show()
```

```python
# Laso Regression
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(x_train,y_train)
y_pred1=lasso.predict(x_test)
mse=mean_squared_error(y_test,y_pred1)
mae=mean_absolute_error(y_test,y_pred1)
score=r2_score(y_test,y_pred1)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
print(score)
plt.scatter(y_test,y_pred1)
plt.show()
```

```python
# Cross Validation Lasso
from sklearn.linear_model import LassoCV
cv=LassoCV(cv=5)
cv.fit(x_train,y_train)
```

```python
y_pred2=cv.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred2)
mae=mean_absolute_error(y_test,y_pred2)
score=r2_score(y_test,y_pred2)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
print(score)
plt.scatter(y_test,y_pred2)
plt.show()
```

```python
# Ridge Regression model
from sklearn.linear_model import Ridge
r=Ridge()
r.fit(x_train,y_train)
```

```python
y_pred3=r.predict(x_test)
mse=mean_squared_error(y_test,y_pred3)
mae=mean_absolute_error(y_test,y_pred3)
score=r2_score(y_test,y_pred3)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
print(score)
plt.scatter(y_test,y_pred3)
plt.show()

```

```python
# Cross Validation Ridge
from sklearn.linear_model import RidgeCV
cv2=RidgeCV(cv=5)
cv2.fit(x_train,y_train)
```

```python
y_pred4=cv2.predict(x_test)
mse=mean_squared_error(y_test,y_pred4)
mae=mean_absolute_error(y_test,y_pred4)
score=r2_score(y_test,y_pred4)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
print(score)
plt.scatter(y_test,y_pred4)
plt.show()

```

```python
# Elasticnet Regression
from sklearn.linear_model import ElasticNet
e=ElasticNet()
e.fit(x_train,y_train)
```

```python
y_pred5=e.predict(x_test)
mse=mean_squared_error(y_test,y_pred5)
mae=mean_absolute_error(y_test,y_pred5)
score=r2_score(y_test,y_pred5)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
print(score)
plt.scatter(y_test,y_pred5)
plt.show()


```

```python
y_pred6=cv3.predict(x_test)
mse=mean_squared_error(y_test,y_pred6)
mae=mean_absolute_error(y_test,y_pred6)
score=r2_score(y_test,y_pred6)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
print(score)
plt.scatter(y_test,y_pred6)
plt.show()
```
